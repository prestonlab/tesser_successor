```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tesser import util

rsa_dir = '/Users/morton/Dropbox/work/tesser/batch/rsa/brsa'
rois = ['b_hip_ant', 'b_hip_body', 'b_hip_tail']
```

```python
subjects = util.subj_list()
rdm = {}
for roi in rois:
    rdm[roi] = [
        np.load(os.path.join(rsa_dir, roi, f'sub-{subject}_brsa.npz'))['C']
        for subject in subjects
    ]
```

```python
len(subjects)
```

```python
mrdm = {
    roi: np.mean(np.dstack([rdm[roi][subject] for subject in range(len(subjects))]), axis=2)
    for roi in rois
}
```

```python
scale = 10
fig, ax = plt.subplots(1, 3, figsize=(1 * scale, 3 * scale))
for i, roi in enumerate(rois):
    ax[i].pcolor(mrdm[roi], vmin=-1, vmax=1, cmap='RdBu_r')
    ax[i].set_title(roi)
    ax[i].set_aspect(1)
    ax[i].invert_yaxis()
```

```python
roi = 'b_hip_ant'
n_row = 8
n_col = 5
scale = 2.5
fig, ax = plt.subplots(8, 5, figsize=(5 * scale, 8 * scale))
n = 0
for i in range(n_row):
    for j in range(n_col):
        if n > (len(subjects) - 1):
            break
        ax[i, j].pcolor(rdm[roi][n], vmin=-1, vmax=1, cmap='RdBu_r')
        ax[i, j].set_aspect(1)
        ax[i, j].invert_yaxis()
        n += 1
```
