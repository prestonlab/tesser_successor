```python
#normality checks
stats.shapiro(bias_behavior['Overall_Bias'])
stats.shapiro(parse_behavior['Parse_Diff'])
stats.shapiro(group_behavior['within_dist'])
```

```python
w = sns.jointplot(neural['b_hip_ant']['within'], bias_behavior['Overall_Bias'], kind='reg', color=sns.xkcd_rgb["pea green"]);
w = w.annotate(stats.pearsonr)
plt.ylabel('parsing performance')
#plt.ylim(0, 1.01)
plt.xlabel('Within-Community Similarity - Fisher''s z (Bilateral Anterior Hippocampus)')
#plt.xlim(5, 35.01)
plt.show()
#w.savefig('b_hip_ant_withinsim_Induct.png', dpi = 300)
#w.savefig('b_hip_ant_withinsim_Induct.pdf', transparent=True)
```
