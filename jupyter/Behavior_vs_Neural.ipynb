```python
def log_reg_df(sub_num, roi, data_dir):
    sub_induct = util.load_induct_subject(data_dir, sub_num)
    lsmeans_items = pd.read_csv((os.path.join(data_dir + '/scrambled_betas/averaged/' + roi + '/' + str(sub_num) + '_' + roi + '_LSMeans_scrambled_ItemSim.csv')), header=None)

    cue_opt1_sim_tog = []
    cue_opt2_sim_tog = []
    chosen_sim_tog = []
    other_sim_tog = []
    
    for s in range(len(sub_induct)):
        cue_index = sub_induct.cue[s]
        opt1_index = sub_induct.opt1[s]
        opt2_index = sub_induct.opt2[s]
    
        cue_opt1_sim = lsmeans_items.loc[cue_index, opt1_index]
        cue_opt1_sim_tog.append(cue_opt1_sim)
        cue_opt2_sim = lsmeans_items.loc[cue_index, opt2_index]
        cue_opt2_sim_tog.append(cue_opt2_sim)
    
        choice_made = sub_induct.Resp[s]
        resp_acc = sub_induct.Acc[s]
    
        #gathering choice similarity 
        if choice_made == 1:
            chosen_sim_tog.append(cue_opt1_sim)
            other_sim_tog.append(cue_opt2_sim)
        elif choice_made == 2:
            chosen_sim_tog.append(cue_opt2_sim)
            other_sim_tog.append(cue_opt1_sim)
        else:
            chosen_sim_tog.append('NaN')   
            other_sim_tog.append('NaN')
        
    sub_induct['cue_opt1_sim'] = cue_opt1_sim_tog
    sub_induct['cue_opt2_sim'] = cue_opt2_sim_tog
    sub_induct['chosen_sim'] = chosen_sim_tog
    sub_induct['other_sim'] =  other_sim_tog
    
    return sub_induct
```

```python
ex_roi = 'b_hip_ant'
ex_data_dir = '/Users/athulapudhiyidath/Dropbox/Experiments/tesser_successor/Data'
ex_induct_sim1 = log_reg_df(100, ex_roi, ex_data_dir)
ex_induct_sim2 = log_reg_df(101, ex_roi, ex_data_dir)
```
