```python
#################
#INDUCTIVE INFERENCE TASK PERFORMANCE
#################

all_bias = []
for s in range(0, len(subjs)):
    #loop through subjects    
    subject = subjs[s]
    bias = induct_perf_prop(data_dir, subject)
    np.array(bias).T
    subj_bias = np.append(subject, bias)
    all_bias.append(subj_bias)
overall_infer_bias = pd.DataFrame(all_bias, columns=['subject', 'overall_bias', 'prim_bias', 'bound1_bias', 'bound2_bias'])
overall_infer_bias
```

```python
#################
#GROUPING PERFORMANCE
#################
all_group = []
for s in range(0, len(subjs)):
    #loop through subjects    
    subject = subjs[s]
    within_across_dist = euc_dist_comm_avg(data_dir, subject)
    subj_group = np.append(subject, within_across_dist)
    all_group.append(subj_group)
overall_group_dist = pd.DataFrame(all_group, columns=['subject', 'within_dist', 'across_dist'])
overall_group_dist
```
