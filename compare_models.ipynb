```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tesser import cython_sr
from tesser import cython_fit
from tesser import util
import importlib
```

```python
tran  = pd.read_csv('./Data/results/transition.csv')
comm  = pd.read_csv('./Data/results/community.csv')
gamma_free  = pd.read_csv('./Data/results/gamma_free.csv')
gamma_fixed = pd.read_csv('./Data/results/gamma_fixed.csv')
```

```python
tran.head()
```

# Free Gamma vs Fixed Gamma

```python
fig = plt.figure(figsize=(10,5))
s =-np.arange(10,31)
plt.subplot(121)
plt.scatter(gamma_free.log_like, gamma_fixed.log_like)
plt.plot(s,s,marker='.',c='r')
plt.title("Log Likelihood of Fixed Gamma vs Free Gamma")
plt.xlabel("Free Gamma Matrix")
plt.ylabel("Fixed Gamma Matrix")
plt.xlim(-30,-10)
plt.ylim(-30,-10)


plt.subplot(122)
plt.scatter(gamma_free.subject, gamma_free.log_like,
            label = 'Free Gamma',c='g')
plt.scatter(gamma_fixed.subject, gamma_fixed.log_like,
            label = 'Fixed Gamma',c='orange')
# plt.scatter(s,s,marker='.')
plt.title("Log Likelihood of Subjects")
plt.xlabel("Subjects")
plt.legend()
# plt.xlim(-30,-10)
plt.ylim(-30,-10)
# plt.figsize(10,10)
```
